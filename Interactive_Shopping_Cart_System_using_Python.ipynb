# **Interactive Shopping Cart System using Python**

```python
total = 0
count = 0

while True:
    price = float(input("Enter item price (0 to stop): "))

    if price == 0:
        break

    if price < 0:
        print("Invalid price! Please enter again.")
        continue

    total += price
    count += 1

# Apply discount
if total >= 1000:
    discount = total * 0.20
elif total >= 500:
    discount = total * 0.10
elif total >= 200:
    discount = total * 0.05
else:
    discount = 0

final_total = total - discount

# Bill Summary
print("\n----- BILL SUMMARY -----")
print(f"Total items: {count}")
print(f"Subtotal: ₹{total:.2f}")
print(f"Discount: ₹{discount:.2f}")
print(f"Final Total: ₹{final_total:.2f}")
```
